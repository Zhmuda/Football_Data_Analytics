#Календрь по сезону

```python
import soccerdata as sd
import pandas as pd

# Создаём объект FBref для АПЛ сезона 2024-2025
fbref = sd.FBref(leagues="ENG-Premier League", seasons=2024)

# Загружаем календарь матчей всей лиги
epl_schedule = fbref.read_schedule()
print("Общий календарь матчей АПЛ:")
display(epl_schedule.head())
display(epl_schedule.columns)
```

# Оставляем команду, чтобы найти ее game_id

```python
# Оставляем только матчи МанСити
man_city_games = epl_schedule[(epl_schedule["home_team"] == "Manchester City") |
                              (epl_schedule["away_team"] == "Manchester City")]

# Выводим первые строки и game_id
print("Матчи МанСити из epl_schedule с game_id:")
display(man_city_games[["date", "home_team", "away_team", "score", "game_id"]].head(11))
```

## Можем посомтреть расписание для этой команды

```python
# Получаем расписание матчей МанСити
team_match_stats = fbref.read_team_match_stats(stat_type="schedule", team="Manchester City")
print("Расписание и результаты матчей МанСити:")
display(team_match_stats.head())
display(team_match_stats.columns)
```

## Смотрим уникальные события в его играх

```python
# Получаем ID матчей МанСити
match_ids = man_city_games["game_id"].dropna().tolist()
print("Список матчей:", match_ids)

unique_event_types = set()  # Множество для уникальных событий

# Проверяем, есть ли доступные матчи
if match_ids:
    for match_id in match_ids[:10]:  # Берём первые 10 матчей
        events = fbref.read_events(match_id=match_id)

        if not events.empty and "event_type" in events.columns:
            unique_event_types.update(events["event_type"].unique())  # Добавляем события в множество

        print(f"Обработан матч {match_id}")

    print("Уникальные типы событий во всех матчах:")
    print(unique_event_types)
else:
    print("Нет доступных матчей для загрузки событий.")
```

#3агружаем общую статистику игроков по команде

```python
# Загружаем сезонную статистику игроков
player_stats = fbref.read_player_season_stats(stat_type='standard')
display(player_stats.head(1))

# Фильтруем игроков по команде в индексе
team_name = 'Arsenal'
filtered_player_stats = player_stats.loc[player_stats.index.get_level_values('team') == team_name]

# Преобразуем MultiIndex в простые названия колонок
filtered_player_stats.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in filtered_player_stats.columns]


display(filtered_player_stats.columns)
display(filtered_player_stats.head())
```

#3агружаем статистику ударов игроков по команде

```python
# Загружаем сезонную статистику игроков
player_stats = fbref.read_player_season_stats(stat_type='shooting')

team_name = 'Arsenal'
filtered_player_stats = player_stats.loc[player_stats.index.get_level_values('team') == team_name]

# Преобразуем MultiIndex в простые названия колонок
filtered_player_stats.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in filtered_player_stats.columns]

# Сбрасываем индекс, чтобы player стал колонкой
filtered_player_stats = filtered_player_stats.reset_index()

# Просмотрим первые строки данных
display(filtered_player_stats.head())
```

## Игроки с наибольшей эффективностью в завершении атак

```python
# Игроки с наибольшей эффективностью в завершении атак
top_goal_efficiency = filtered_player_stats.sort_values(by='Standard_G/Sh', ascending=False).head(10)
display(top_goal_efficiency[['player', 'Standard_G/Sh']])
```

## Игроки с высокой точностью ударов


```python
# Игроки с высокой точностью ударов
top_accuracy = filtered_player_stats[filtered_player_stats['Standard_SoT'] / filtered_player_stats['Standard_Sh'] > 0.5]
display(top_accuracy[['player', 'Standard_SoT', 'Standard_Sh']])
```

## Игроки с дальними ударами и голам с дистанции > 20 метров

```python
# Добавляем колонку с количеством голов с дальних ударов
filtered_player_stats['Long_Dist_Gls'] = filtered_player_stats['Standard_Gls'] * (filtered_player_stats['Standard_Dist'] > 20)

# Игроки с самыми дальними ударами, теперь с количеством голов с дальних ударов
top_long_shots = filtered_player_stats[filtered_player_stats['Long_Dist_Gls'] > 0].sort_values(by='Standard_Dist', ascending=False).head(5)

display(top_long_shots[['player', 'Standard_Dist', 'Long_Dist_Gls']])
```

## Игроки, выполняющие много штрафных ударов
## Игроки, выполняющие пенальти

```python
# Игроки, выполняющие много штрафных ударов
top_free_kick_takers = filtered_player_stats[filtered_player_stats['Standard_FK'] > 0].sort_values(by='Standard_FK', ascending=False).head(10)
display(top_free_kick_takers[['player', 'Standard_FK']])

# Игроки с наибольшим количеством забитых пенальти
top_penalty_takers = filtered_player_stats[filtered_player_stats['Standard_PK'] > 0].sort_values(by='Standard_PK', ascending=False).head(10)
display(top_penalty_takers[['player', 'Standard_PK']])
```

## Среднее количество голов и ожидаемых голов по позициям

```python
# Создаем маппинг позиций к основным категориям
position_mapping = {
    'GK': 'GK', 'DF': 'DF', 'FB': 'DF', 'LB': 'DF', 'RB': 'DF', 'CB': 'DF',
    'DM': 'MF', 'CM': 'MF', 'LM': 'MF', 'RM': 'MF', 'WM': 'MF', 'AM': 'MF','MF': 'MF',
    'LW': 'FW', 'RW': 'FW', 'FW': 'FW'
}

# Применяем маппинг к столбцу позиций
filtered_player_stats['Main_Pos'] = filtered_player_stats['pos_'].map(position_mapping)

# Группируем по основным позициям и считаем средние значения голов и xG
position_groups = filtered_player_stats.groupby('Main_Pos')[['Standard_Gls', 'Expected_xG']].mean()

# Выводим только DF, MF, FW
display(position_groups.loc[['DF', 'MF', 'FW']])

```

## Игроки с наибольшей точностью ударов

```python
# Добавляем столбец точности ударов (процент ударов в створ)
filtered_player_stats['Shot_Accuracy'] = filtered_player_stats['SoT/90'] / filtered_player_stats['Sh/90']
filtered_player_stats.loc[filtered_player_stats['Sh/90'] == 0, 'Shot_Accuracy'] = 0  # Если нет ударов, точность = 0

# Убираем игроков с NaN
filtered_player_stats = filtered_player_stats.dropna(subset=['Sh/90', 'SoT/90', 'Expected_npxG'])

# Игроки с наибольшим количеством ударов за 90 минут
top_shooters = filtered_player_stats.sort_values(by='Sh/90', ascending=False)[['player', 'Sh/90', 'SoT/90', 'Shot_Accuracy', 'Expected_npxG']]

# Игроки с наибольшей точностью ударов (у кого хотя бы 2 удара)
top_accurate_shooters = filtered_player_stats[filtered_player_stats['Sh/90'] > 1].sort_values(by='Shot_Accuracy', ascending=False)[['player', 'Sh/90', 'SoT/90', 'Shot_Accuracy', 'Expected_npxG']]
display(top_accurate_shooters.head(5))
```

## Игроки с наибольшим npxG (без учета пенальти)

```python
# Игроки с наибольшим npxG (без учета пенальти)
top_npxG = filtered_player_stats.sort_values(by='Expected_npxG', ascending=False)[['player', 'Sh/90', 'SoT/90', 'Shot_Accuracy', 'Expected_npxG']]
display(top_npxG.head(5))
```

# Загружаем сезонную статистику пасов игроков


```python
# Загружаем сезонную статистику игроков
player_pas_stats = fbref.read_player_season_stats(stat_type='passing')

team_name = 'Arsenal'
filtered_player_pas_stats = player_pas_stats.loc[player_pas_stats.index.get_level_values('team') == team_name]

# Преобразуем MultiIndex в простые названия колонок
filtered_player_pas_stats.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in filtered_player_pas_stats.columns]

# Сбрасываем индекс, чтобы player стал колонкой
filtered_player_pas_stats = filtered_player_pas_stats.reset_index()

# Просмотрим первые строки данных
display(filtered_player_pas_stats.head(5))
display(filtered_player_pas_stats.columns)
```

##1. Общая точность и объем передач

```python
top_passers = filtered_player_pas_stats.sort_values(by='Total_Att', ascending=False).head(5)
display(top_passers[['player', 'Total_Att', 'Total_Cmp%']])
```

##2. Разделение по дальности передач


```python
top_short_passers = filtered_player_pas_stats[filtered_player_pas_stats['Short_Att'] > 10].sort_values(by='Short_Cmp%', ascending=False).head(5)

top_medium_passers = filtered_player_pas_stats[filtered_player_pas_stats['Medium_Att'] > 10].sort_values(by='Medium_Cmp%', ascending=False).head(5)

top_long_passers = filtered_player_pas_stats[filtered_player_pas_stats['Long_Att'] > 10].sort_values(by='Long_Cmp%', ascending=False).head(5)

display(top_short_passers[['player', 'Short_Cmp%']])
display(top_medium_passers[['player', 'Medium_Cmp%']])
display(top_long_passers[['player', 'Long_Cmp%']])
```

##3. Лидеры по ассистам и xA


```python
top_assist_providers = filtered_player_pas_stats.sort_values(by='Ast_', ascending=False).head(5)

display(top_assist_providers[['player', 'Ast_', 'Expected_xA']])

#Ast_ > Expected_xA → игрок "переисполняет", значит, партнеры хорошо реализуют его передачи.
#Ast_ < Expected_xA → игрок создает моменты, но партнеры плохо завершают атаки.
```

##4. Лидеры по ключевым передачам и продвижению мяча


```python
top_key_passers = filtered_player_pas_stats.sort_values(by='KP_', ascending=False).head(5)
top_final_third = filtered_player_pas_stats.sort_values(by='1/3_', ascending=False).head(5)
top_penalty_area = filtered_player_pas_stats.sort_values(by='PPA_', ascending=False).head(5)
top_progressive_passers = filtered_player_pas_stats.sort_values(by='PrgP_', ascending=False).head(5)

display(top_key_passers[['player', 'KP_']])
display(top_final_third[['player', '1/3_']])
display(top_penalty_area[['player', 'PPA_']])
display(top_progressive_passers[['player', 'PrgP_']])

#KP_ → ключевые передачи (под удары)
#1/3_ → пасы в финальную треть → важно для распасовщиков.
#PPA_ → передачи в штрафную → круто для созидателей моментов.
#Чем выше PrgP_, тем больше игрок продвигает мяч вперед.
```

##6. Средние показатели по позициям

##1. Кто больше всего делает креативные передачи?

* Through Balls (Pass Types_TB) – передача между защитниками
* Switches (Pass Types_Sw) – смена фланга
* Crosses (Pass Types_Crs) – навесы

```python
top_through_ball_players = filtered_player_pas_type_stats.sort_values(by='Pass Types_TB', ascending=False).head(5)
top_switch_players = filtered_player_pas_type_stats.sort_values(by='Pass Types_Sw', ascending=False).head(5)
top_cross_players = filtered_player_pas_type_stats.sort_values(by='Pass Types_Crs', ascending=False).head(5)

display(top_through_ball_players[['player', 'Pass Types_TB']])
display(top_switch_players[['player', 'Pass Types_Sw']])
display(top_cross_players[['player', 'Pass Types_Crs']])

#Игроки с наибольшим количеством сквозных передач (TB) – классические плеймейкеры.
#Много смен фланга (Sw) – защитники или полузащитники, часто меняющие направление атаки.
#Навесы (Crs) – крайние защитники и вингеры, активно работающие в ширине.
```
