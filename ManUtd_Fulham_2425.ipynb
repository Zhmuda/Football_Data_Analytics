#1. Анализ эффективности реализации моментов (xG vs. Goals)
Идея: сравнить ожидаемые голы (xG) и фактические голы (goals) для команд.

Выявим команды, которые реализуют больше моментов, чем предполагает xG (эффективные нападающие).
Найдём команды, которые забивают меньше, чем ожидается (проблемы с реализацией).
Метрики: home_xg, home_goals, away_xg, away_goals из team_match_stats.

```python
team_match_stats["home_xg_diff"] = team_match_stats["home_goals"] - team_match_stats["home_xg"]
team_match_stats["away_xg_diff"] = team_match_stats["away_goals"] - team_match_stats["away_xg"]

xg_analysis = team_match_stats.groupby("home_team")[["home_xg_diff"]].mean()
xg_analysis = xg_analysis.merge(team_match_stats.groupby("away_team")[["away_xg_diff"]].mean(),
                                left_index=True, right_index=True)

xg_analysis["total_xg_diff"] = (xg_analysis["home_xg_diff"] + xg_analysis["away_xg_diff"]) / 2
xg_analysis = xg_analysis.sort_values("total_xg_diff", ascending=False)

xg_analysis.head(10)  # Топ-10 команд по превышению xG
```

Если команда стабильно превышает xG, это может говорить о:
* Отличной реализации моментов (например, хорошие нападающие).
* Качестве ударов (бьют из хороших позиций).
* Везении (это может не сохраниться в будущем).

Если команда забивает меньше xG, возможны причины:
* Проблемы с нападающими (плохая реализация).
* Неудачные стечения обстоятельств (может улучшиться со временем).
* Вратари соперников играют выше среднего уровня.

#2. Кто лучший ассистент? (xA vs Assists)
Идея: найти игроков, у которых xA (ожидаемые ассисты) выше фактических ассистов — возможно, партнёры плохо реализуют их передачи.
Также можно найти игроков с высоким assists - xa — они делают меньше передач, но их партнёры забивают после них.

Метрики: assists, xa из player_season_stats.

```python
player_season_stats["assist_efficiency"] = player_season_stats["assists"] - player_season_stats["xa"]
top_assisters = player_season_stats.sort_values("assist_efficiency", ascending=False)[["player_id", "assists", "xa", "assist_efficiency"]]

top_assisters.head(10)  # Топ-10 игроков, у которых ассистов больше ожидаемых
```

* Игроки с высоким assist_efficiency могут просто иметь удачливых партнёров.
* Игроки с низким — возможно, недооценены, так как создают много моментов, но партнёры не реализуют их передачи.

#3. Давление на соперника: PPDA и Deep Completions
Идея: PPDA (Passes per Defensive Action) показывает, насколько активно команда прессингует соперника.
Deep completions — успешные передачи в опасную зону.
Сравним их, чтобы понять, какие команды агрессивно играют в обороне и создают давление.

Метрики: home_ppda, away_ppda, home_deep_completions, away_deep_completions из team_match_stats.

```python
ppda_deep = team_match_stats.groupby("home_team")[["home_ppda", "home_deep_completions"]].mean()
ppda_deep = ppda_deep.merge(team_match_stats.groupby("away_team")[["away_ppda", "away_deep_completions"]].mean(),
                            left_index=True, right_index=True)

ppda_deep["avg_ppda"] = (ppda_deep["home_ppda"] + ppda_deep["away_ppda"]) / 2
ppda_deep["avg_deep_completions"] = (ppda_deep["home_deep_completions"] + ppda_deep["away_deep_completions"]) / 2
ppda_deep = ppda_deep.sort_values("avg_ppda")

ppda_deep.head(20)  # Топ-20 команд с самым низким PPDA (сильный прессинг)
```

* Низкий avg_ppda → команда активно прессингует.
* Высокий avg_deep_completions → команда часто проникает в опасную зону.

* Если у команды высокий deep completions, значит, они создают много атак через контроль мяча.

#4. Карта ударов (Shot Map)
Идея: визуализируем удары по полю, чтобы увидеть, из каких зон чаще всего бьют игроки.

Метрики: location_x, location_y, xg из shot_events.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Загружаем данные
df_shots = shot_events.copy()

# Создаём объект поля (половина поля)
pitch = Pitch(pitch_type='statsbomb', line_color='white', pitch_color='grass',
              half=True, stripe=True)

# Фильтруем удары команды
team_id = 228
team_name = 'Man Utd'
df_team_shots = df_shots[df_shots['team_id'] == team_id].copy()

# Приведение координат к половине поля
df_team_shots['location_x'] *= 120
df_team_shots['location_y'] = (1 - df_team_shots['location_y']) * 80

# Создаём фигуру и оси
fig, ax = pitch.draw(figsize=(10, 6))

# Рисуем удары
sc = ax.scatter(
    df_team_shots['location_x'],
    df_team_shots['location_y'],
    c=df_team_shots['xg'],
    cmap='coolwarm',
    edgecolors='black',
    s=df_team_shots['xg'] * 500
)

# Добавляем цветовую шкалу
plt.colorbar(sc, label='xG (Ожидаемый гол)')

# Заголовок
plt.title(f'Карта ударов {team_name} в матче', fontsize=14)

# Показываем график
plt.show()
```

* Если команда бьёт в основном из-за штрафной с низким xG → неэффективная атака.
* Если команда создаёт моменты в центре штрафной с высоким xG → качественное нападение.
